# TC-data

```python
data = pd.read_csv('E:/Work/研二/1226HYY数据/HYY/下车时间匹配/下车时间匹配成功.csv',low_memory=False)
data['enstation_time'] = pd.to_datetime(data['enstation_time'])
data['exstation_time'] = pd.to_datetime(data['exstation_time'],format='mixed')
data
```

```python
def haversine_distance(df):  #距离计算
    import math
    R = 6371
    dLon = math.radians(df['lon2'] - df['lon1'])
    dLat = math.radians(df['lat2'] - df['lat1'])
    lat1 = math.radians(df['lat1'])
    lat2 = math.radians(df['lat2'])
    a = math.sin(dLat / 2) ** 2 + \
        math.sin(dLon / 2) ** 2 * math.cos(df['lat1']) * math.cos(df['lat2'])
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    dis = R * c
    return dis
```

```python
data['distance'] = data.apply(haversine_distance,axis=1)
```

```python
# data.to_csv('data/RD_distance.csv',index=False)
```

```python
data
```

```python
bins = [i for i in range(0,25)]
```

```python
out = pd.DataFrame(columns=['区间','amount'])
out_1 = []
out_2 = []
tem_1 = data
for i in bins:
    tem_2 = tem_1[tem_1['distance'].between(i,i+1,inclusive='left')].reset_index(drop=True)
    out_1.append(f'[{i},{i+1})')
    out_2.append(len(tem_2["card_no"].unique()))
    print(f'区间：[{i},{i+1})    {len(tem_2["card_no"].unique())}')
tem_2 = tem_1[tem_1['distance'].between(i,i+10000,inclusive='left')].reset_index(drop=True)
out_1.append('大于100')
out_2.append(len(tem_2["card_no"].unique()))
out['区间'] = out_1
out['amount'] = out_2
print(f'区间：大于{bins[-1] + 1}    {len(tem_2["card_no"].unique())}')
print(f'总共有{len(tem_1["card_no"].unique())}人')
```

```python
out
```

```python
out_A = out
```

```python
def Redirection_1(Dataframe,col_1,col_2):
    redirection = {}
    df = Dataframe[[col_1,col_2]]
    for i in (range(len(Dataframe))):
        redirection[Dataframe[col_1].iloc[i]] = Dataframe[col_2].iloc[i]
    return redirection
```
