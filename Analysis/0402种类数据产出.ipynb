# XGBoost

```python
def out_p(data,out_name):
    timeframe = []
    length = []
    for i in range(24):
        tem = data[(data['exstation_time'].dt.time >= pd.to_datetime(str(i) + ':00:00').time()) & 
                    (data['exstation_time'].dt.time <= pd.to_datetime(str(i) + ':59:59').time())]
        timeframe.append(f'{str(i)}:00:00-{str(i)}:59:59')
        length.append(len(tem))
        print(f'时间范围:{str(i)}:00:00-{str(i)}:59:59    {len(tem)}')
    out_dataframe = pd.DataFrame(data={'时间段': timeframe, '数量': length})
    out_dataframe.to_csv('data/' + out_name + '.csv',index=False)
```

```python
data = pd.read_csv('0402XG下车时间五.csv',low_memory=False)
data['enstation_time'] = pd.to_datetime(data['enstation_time'])
data['exstation_time'] = pd.to_datetime(data['exstation_time'],format='mixed')
data = data[~data['exstation_time'].isna()]
data = data[data['enstation_time'] != data['exstation_time']]
data
```

```python
out_p(data_laodeng,'XGBoost_eld')
out_p(data_O,'XGBoost_o')
out_p(data_stu,'XGBoost_s')
```
