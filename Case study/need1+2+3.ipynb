```python
def in2F(Dataframe_1,Dataframe_2):
    out = []
    lon1 = Dataframe_1['lon1']
    lat1 = Dataframe_1['lat1']
    line_name = Dataframe_1['line_name_previous']
    orientation = Dataframe_1['orientation_previous']
    try:
        bus_line_rule = (Dataframe_2['line_name'] == line_name) & \
                        (Dataframe_2['orientation'] == orientation)
        use_bus = Dataframe_2[bus_line_rule]
        use_bus['dis'] = use_bus.apply(haversine_distance, args=[lon1, lat1], axis=1)
        min_index = use_bus['dis'].idxmin()
        min_row = use_bus.loc[min_index]
        out.append(min_row['gcj02_lng'])
        out.append(min_row['gcj02_lat'])
        out.append(min_row['stop_name'])
        return out
    except:
        return out
        print('可排除误差')

def haversine_distance(Dataframe, lon2, lat2):
    import math
    R = 6371
    dLon = math.radians(lon2 - Dataframe['gcj02_lng'])
    dLat = math.radians(lat2 - Dataframe['gcj02_lat'])
    lat1 = math.radians(Dataframe['gcj02_lat'])
    lat2 = math.radians(lat2)
    a = math.sin(dLat / 2) ** 2 + \
        math.sin(dLon / 2) ** 2 * math.cos(lat1) * math.cos(lat2)
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    dis = R * c
    return dis
```
