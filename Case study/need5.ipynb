```python
def haversine_distance(row, lon2, lat2):  #距离计算
    import math
    R = 6371
    dLon = math.radians(lon2 - row['lon1_previous'])
    dLat = math.radians(lat2 - row['lat1_previous'])
    lat1 = math.radians(row['lat1_previous'])
    lat2 = math.radians(lat2)
    a = math.sin(dLat / 2) ** 2 + \
        math.sin(dLon / 2) ** 2 * math.cos(lat1) * math.cos(lat2)
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    dis = R * c
    return dis
```
