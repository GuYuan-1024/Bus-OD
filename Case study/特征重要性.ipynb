```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl


# 特征及其重要性
features = [
    'Latitude', 'Longitude', 'Card ID', 'Bus ID', 'Minute',
    'Line name', 'Hour', 'Day', 'Line length', 'Card type',
    'Direction', 'Second', 'Stop ID'
]
importances = [
    0.146, 0.122, 0.103, 0.095, 0.070,
    0.056, 0.056, 0.055, 0.051, 0.043,
    0.036, 0.028, 0.010
]

# 构建DataFrame
df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

# 按重要性排序
df = df.sort_values(by='Importance', ascending=True)


# 设置风格
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6),dpi=300)
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
# 绘制水平条形图
bars = plt.barh(df['Feature'], df['Importance'], color='gray', edgecolor='black')

# 添加百分比标签
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.002, bar.get_y() + bar.get_height()/2,
             f'{width*100:.1f}%', 
             va='center', 
             ha='left', 
             fontsize=10,
             fontname='Times New Roman')  # 确保标签也使用新罗马
plt.xticks(fontname='Times New Roman')
plt.yticks(fontname='Times New Roman')

# 标题与坐标轴
plt.xlabel('Feature Importance',fontname='Times New Roman')  # Feature Importance
plt.ylabel('Feature Name',fontname='Times New Roman')    # Feature Name
# plt.title('特征重要性分析') # Feature Importance Analysis


plt.tight_layout()
plt.savefig('D:/Work/研二/0607小论文/0910/图/Figure 3.tiff',format="tiff",dpi=300)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl

# 特征及其重要性
features = [
    'Latitude', 'Longitude', 'Card ID', 'Bus ID', 'Minute',
    'Line name', 'Hour', 'Day', 'Line length', 'Card type',
    'Direction', 'Second', 'Stop ID'
]
importances = [
    0.146, 0.122, 0.103, 0.095, 0.070,
    0.056, 0.056, 0.055, 0.051, 0.043,
    0.036, 0.028, 0.010
]

# 构建DataFrame
df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

# 按重要性排序
df = df.sort_values(by='Importance', ascending=True)

# 设置风格
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6), dpi=300)
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来显示中文
plt.rcParams['axes.unicode_minus'] = False

# 使用Seaborn调色板
colors = sns.color_palette("Set2", len(df))

# 绘制彩色条形图
bars = plt.barh(df['Feature'], df['Importance'], color=colors, edgecolor='black')

# 添加百分比标签
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.002, bar.get_y() + bar.get_height()/2,
             f'{width*100:.1f}%', 
             va='center', 
             ha='left', 
             fontsize=10,
             fontname='Times New Roman')

# 设置字体
plt.xticks(fontname='Times New Roman')
plt.yticks(fontname='Times New Roman')

# 坐标轴标签
plt.xlabel('Feature Importance', fontname='Times New Roman')
plt.ylabel('Feature Name', fontname='Times New Roman')

plt.tight_layout()
plt.show()
```
